```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import DistilBertTokenizer, DistilBertModel
import torch
import re
from transformers import AutoTokenizer, AutoModelForSequenceClassification
# from nltk.corpus import stopwords
```

```python
df = pd.read_csv("/content/final_dataset.csv")
print("Dataset loaded! First 5 rows:")
print(df.head())
```

```python
# Clean text function
def clean_text(text):
  text = text.lower()
  text = re.sub(r'http\S+|https\S+', "", text, flags=re.MULTILINE)

  # Remove puctuation/special chars
  text = re.sub(r'[^\w\s]', "", text)
  # stop_words = set(stopwords.words("english"))

  # word_tokens = text.split()
  # text = " ".join([word for word in word_tokens if word not in stop_words])
  # Remove extra spaces
  text = re.sub(r'\s+', " ", text).strip()

  return text
```

```python
df['Content'] = df['Content'].apply(clean_text)
print(df.head())
```

```python
df.to_csv("cleaned_dataset.csv", index=False)

from google.colab import files

files.download("cleaned_dataset.csv")
```

```python
df = pd.read_csv("cleaned_dataset.csv")
print(df.head())
```

```python
# Handle imbalance with SMOTE
x = df["Content"].tolist()
y = df["Label"].tolist()

# Convert text to TF-IDF vectors
tfidf = TfidfVectorizer(max_features=5000)
x_tfidf = tfidf.fit_transform(x).toarray()

y = df["Label"].values

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x_tfidf, y)
x_balanced = [x[0] for x in x_resampled]

y_balanced = y_resampled
```

```python
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = DistilBertModel.from_pretrained("distilbert-base-uncased")
```

```python
tokenizer =  AutoTokenizer.from_pretrained("Davlan/afro-xlmr-large")
model = AutoModelForSequenceClassification.from_pretrained("Davlan/afro-xlmr-large", num_labels=2)
```

```python
# Convert text to embeddings
def get_embeddings(text):
  inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
  with torch.no_grad():
    outputs = model(**inputs)

  return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

x_embeddings = [get_embeddings(claim) for claim in x_balanced]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_resampled,
    y_resampled,
    test_size=0.2,
    random_state=42,
    stratify=y_resampled
)

print(f"Train shape: {x_train.shape}, Test shape: {x_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, class_weight="balanced")
clf.fit(x_train, y_train)
```

```python
accuracy = clf.score(x_test, y_test)
print(f"Accuracy: {accuracy: .2f}")
```

```python
y_pred = clf.predict(x_test)
```

```python
y_probs = clf.predict_proba(x_test)[:, 1]
y_pred = (y_probs > 0.6).astype(int)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
